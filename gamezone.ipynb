```python
month_revenue_trend = df_normal.resample('ME', on = 'PURCHASE_TS')["USD_PRICE"].sum()
month_revenue_trend.plot(marker = 'o', figsize = (10,6))
plt.title("Revenue Trend", fontsize = 16, fontweight = 'bold')
plt.xlabel("")
plt.yticks([100000, 200000, 300000, 400000, 500000], ['100K', '200K', '300K', '400K', '500K'])
```

```python
MoM_growth = month_revenue_trend.pct_change() * 100
MoM_growth
```

```python
MoM_growth.plot(label = "MoM Growth")
plt.axhline(0, color = 'black', linestyle = '--')
plt.xlabel("")
plt.ylabel("Change %")
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(month_revenue_trend.index, month_revenue_trend, alpha=0.6, label='Monthly Revenue', width=20)
ax1.set_ylabel('Total Revenue (USD)', fontsize=12)
ax1.set_yticks([100000, 200000, 300000, 400000, 500000])
ax1.set_yticklabels(['100K', '200K', '300K', '400K', '500K'])

ax2 = ax1.twinx()  
ax2.plot(MoM_growth.index, MoM_growth, color='orange', marker='o', linewidth=2, label='MoM Growth %')
ax2.set_ylabel('Growth Rate (%)', fontsize=12, color='orange')
ax2.tick_params(axis='y', labelcolor='r')

plt.title("Revenue starts to climb in April 2020, reaching its peak in January 2021 then declining.", fontsize=16, fontweight='bold')
ax1.grid(axis='y', linestyle='--', alpha=0.5)
fig.tight_layout()
plt.show()
```
